```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# 📌 Update this path with your macOS username
file_path = "/Users/anu/Downloads/Tweet.xlsx"  # Change 'your_username'

# Load the Excel file
df = pd.read_excel(file_path)

# Ensure column names match your dataset (adjust if needed)
# Example column names: "Tweet", "Impressions", "URL Clicks", "Profile Clicks", "Hashtag Clicks"

# Filter tweets with impressions > 500
filtered_df = df[df["impressions"] > 500]

# Calculate total clicks for each tweet
filtered_df["Total Clicks"] = filtered_df[["url clicks", "user profile clicks", "hashtag clicks"]].sum(axis=1)

# Aggregate total clicks across categories
aggregated_clicks = filtered_df[["url clicks", "user profile clicks", "hashtag clicks"]].sum()

# 📊 Generate Pie Chart for overall proportions
plt.figure(figsize=(8, 8))
plt.pie(
    aggregated_clicks,
    labels=aggregated_clicks.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=["#66b3ff", "#99ff99", "#ffcc99"]
)
plt.title("Proportion of Total Clicks (Tweets with >500 Impressions)")
plt.show()

# 🔍 Interactive Drill-Down Visualization with Plotly
fig = px.sunburst(
    filtered_df.melt(id_vars=["Tweet"], value_vars=["url clicks", "user profile clicks", "hashtag clicks"]),
    path=["Tweet", "variable"],
    values="value",
    color="variable",
    title="Drill-Down: Click Distribution per Tweet",
    color_discrete_map={"url clicks": "#66b3ff", "user profile clicks": "#99ff99", "hashtag clicks": "#ffcc99"}
)

fig.show()
```
